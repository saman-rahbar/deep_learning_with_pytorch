### Salient Features of Pytorch:

- Python first (Integrated with Python) - Unlike Tensorflow
- Pytorch computations run within the Python computation model
- It's impressive, just like Python
- Pytorch can be extended, just like you would extend Python
- Pytorch uses a Dynamic computation graph (Versus Static)
- Computation graph gets created on the fly\
- Pytorch is dynamic. However, Tensorflow is static

*Dynamic (or imperative) means we can write and execute part of the program as you declare them. In static programing you should declare the program before excute the program.

Example of "Static" and "Dynamic/imperative" programing is shown below:

- Pytorch works based on "Define and Run" Model. It means you don't have to writw the whole program to execute and build the graph model. You can write pieces of the model and execute different parts of that (You can execute them as you define them).

- Linear in thought
- Faster prototyping
- Easier to debug in Pytorch. Pytorch does not use any other Virtual Environment.
- It supports pip, Linux, OSX

## Let's Install Pytorch on Linux Environment Now!

We can use anacoda environment to set the system and the virtual environment. You can create a virtual environment using Conda and install python3.6 along with the other packages that you want to instal, including Pytorch.

In order to do that navigate to the terminal on your computer and run the following command line:
- "conda create -n py36 python=3.6 anaconda"

We then activate the virtual environment using the following command:
- "source activate py36"
In case you want to deactivate the source you have to run the following code:
- "source deactivate py36"

We then go to the pytorch.org and based on the config we are using we would get the pip command kto install pytorch in the environment we just created. The command should look something like this:

- "~/anaconda/envs/py36/bin/pip ..."

Yay! We are done :)
