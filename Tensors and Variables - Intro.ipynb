## Refresher on the concepts:

```python
from IPython.display import Image
Image("Concepts.png")
```

- Scaler: As you see in the above imgage, scalers are numbers without any dimension. They can be real number like (1.2) or natural number like 3 or 5. The rank is basically an indicator of how many indexes you will need to locate a number or element.
- Vector: Are ordered list of single numbers. The dimension is the number of elements inside a list.
- Matrix: A rectangular list of numbers. 
- Tensor: Anything beyond rectangular (Matrix)

### Tensors in Pytorch:

- A multi-dimensional matrix containing elements of a single data types
- Part of the torch package
- Instantiated via torch.Tensor group of classes
- Can be sorted on the CPU or GPU
- Operated on via functions available in the torch package, or via class methods
- Interoperable with Numpy array
